```python
def is_prime(n):
    """
    Comprueba si un número n es primo.

    Parámetros:
    n (int): Número a comprobar si es primo.

    Retorna:
    bool: True si el número es primo, False en caso contrario.
    """
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_list(limit):
    """
    Genera una lista de números primos hasta un límite dado.

    Parámetros:
    limit (int): Límite superior (inclusive) para generar números primos.

    Retorna:
    list: Lista de números primos hasta el límite dado.
    """
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def check_palindrome(primes):
    """
    Busca números primos que son palíndromos (se leen igual de adelante hacia atrás).

    Parámetros:
    primes (list): Lista de números primos.

    Retorna:
    list: Lista de números primos palíndromos.
    """
    palindromes = []
    for prime in primes:
        if str(prime) == str(prime)[::-1]:
            palindromes.append(prime)
    return palindromes

def categorize_prime(prime):
    """
    Clasifica los números primos en categorías según su tamaño.

    Parámetros:
    prime (int): Número primo a clasificar.

    Retorna:
    str: Categoría del número primo ("pequeños", "medianos", o "grandes").
    """
    if prime < 10:
        return "pequeños"
    elif 10 <= prime <= 99:
        return "medianos"
    else:
        return "grandes"

def main():
    """
    Función principal para probar las funciones del programa.
    """
    # Probar la función is_prime
    numero = 29
    print(f"¿El número {numero} es primo?: {is_prime(numero)}")

    # Probar la función prime_list
    limite = 100
    primos = prime_list(limite)
    print(f"Lista de números primos hasta {limite}: {primos}")

    # Probar la función check_palindrome
    primos_palindromos = check_palindrome(primos)
    print(f"Números primos palíndromos hasta {limite}: {primos_palindromos}")

    # Probar la función categorize_prime
    for primo in primos:
        categoria = categorize_prime(primo)
        print(f"El número primo {primo} es de categoría: {categoria}")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

```python
# Definir las funciones para cada operación

def suma_total(numeros):
    """
    Calcula la suma de todos los números de la lista.

    Parámetros:
    numeros (list): Lista de números enteros.

    Retorna:
    int: La suma de todos los números.
    """
    return sum(numeros)

def suma_positivos_negativos(numeros):
    """
    Calcula la suma de los números positivos y la suma de los números negativos por separado.

    Parámetros:
    numeros (list): Lista de números enteros.

    Retorna:
    tuple: Suma de números positivos y suma de números negativos.
    """
    suma_positivos = sum(n for n in numeros if n > 0)
    suma_negativos = sum(n for n in numeros if n < 0)
    return suma_positivos, suma_negativos

def max_min(numeros):
    """
    Encuentra el número máximo y mínimo de la lista.

    Parámetros:
    numeros (list): Lista de números enteros.

    Retorna:
    tuple: El número máximo y el número mínimo.
    """
    return max(numeros), min(numeros)

def numeros_unicos(numeros):
    """
    Comprueba si todos los números de la lista son únicos.

    Parámetros:
    numeros (list): Lista de números enteros.

    Retorna:
    bool: True si todos los números son únicos, False en caso contrario.
    """
    return len(numeros) == len(set(numeros))

def mas_positivos_o_negativos(numeros):
    """
    Comprueba si la lista contiene más números positivos o más números negativos.

    Parámetros:
    numeros (list): Lista de números enteros.

    Retorna:
    str: Mensaje que indica si hay más positivos, más negativos o si están equilibrados.
    """
    positivos = len([n for n in numeros if n > 0])
    negativos = len([n for n in numeros if n < 0])
    return "Más números positivos" if positivos > negativos else "Más números negativos" if negativos > positivos else "Igual cantidad de positivos y negativos"

def main():
    """
    Función principal que ejecuta el menú y las funciones correspondientes.
    """
    numeros = list(map(int, input("Ingrese una lista de números separados por espacios: ").split()))

    while True:
        print("\nMenú de opciones:")
        print("1. Calcular la suma de todos los números")
        print("2. Calcular la suma de los números positivos y la suma de los negativos")
        print("3. Encontrar el número máximo y el mínimo de la lista")
        print("4. Comprobar si todos los números son únicos")
        print("5. Comprobar si hay más números positivos o negativos")
        print("6. Salir")

        opcion = input("Elige una opción: ")

        if opcion == '1':
            # Opción 1: Sumar todos los números
            print(f"La suma total de los números es: {suma_total(numeros)}")

        elif opcion == '2':
            # Opción 2: Sumar positivos y negativos por separado
            suma_pos, suma_neg = suma_positivos_negativos(numeros)
            print(f"Suma de números positivos: {suma_pos}")
            print(f"Suma de números negativos: {suma_neg}")

        elif opcion == '3':
            # Opción 3: Encontrar el máximo y el mínimo
            maximo, minimo = max_min(numeros)
            print(f"El número máximo es: {maximo}")
            print(f"El número mínimo es: {minimo}")

        elif opcion == '4':
            # Opción 4: Comprobar si todos los números son únicos
            if numeros_unicos(numeros):
                print("Todos los números en la lista son únicos.")
            else:
                print("Hay números repetidos en la lista.")

        elif opcion == '5':
            # Opción 5: Comprobar si hay más números positivos o negativos
            print(mas_positivos_o_negativos(numeros))

        elif opcion == '6':
            # Opción 6: Salir del programa
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida. Por favor, elija una opción del 1 al 6.")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

```python
# Declarar la lista de tareas como variable global
tareas = []

def add_task(task):
    """
    Agrega una nueva tarea a la lista global de tareas.

    Parámetros:
    task (str): La tarea a agregar.
    """
    global tareas
    tareas.append(task)
    print(f"Tarea '{task}' agregada.")

def remove_task(task):
    """
    Elimina una tarea de la lista global de tareas, si existe.

    Parámetros:
    task (str): La tarea a eliminar.
    """
    global tareas
    if task in tareas:
        tareas.remove(task)
        print(f"Tarea '{task}' eliminada.")
    else:
        print(f"Tarea '{task}' no encontrada.")

def list_tasks():
    """
    Enumera todas las tareas almacenadas en la lista global de tareas.
    """
    global tareas
    if not tareas:
        print("La lista de tareas está vacía.")
    else:
        print("Lista de tareas:")
        for idx, task in enumerate(tareas, 1):
            print(f"{idx}. {task}")

def task_manager():
    """
    Administra las operaciones de agregar, modificar y enumerar tareas.
    """

    def modify_task():
        """
        Modifica la primera tarea de la lista de tareas.
        Esta función está definida dentro de task_manager() para modificar la lista global de tareas.
        """
        global tareas
        if tareas:
            old_task = tareas[0]
            tareas[0] = f"{old_task} (modificada)"
            print(f"Tarea '{old_task}' modificada a '{tareas[0]}'.")
        else:
            print("No hay tareas para modificar.")

    # Modificar la primera tarea
    modify_task()

def main():
    """
    Función principal para probar las operaciones del gestor de tareas.
    """
    # Agregar tareas
    add_task("Comprar leche")
    add_task("Llamar al doctor")
    add_task("Hacer ejercicio")

    # Listar las tareas actuales
    list_tasks()

    # Modificar la primera tarea
    task_manager()

    # Listar nuevamente las tareas después de modificar
    list_tasks()

    # Eliminar una tarea
    remove_task("Hacer ejercicio")

    # Listar las tareas después de eliminar una
    list_tasks()

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

```python
# Definir las funciones para cada operación

def imprimir_cadenas_por_letra(cadenas, letra):
    """
    Imprime cada cadena de la lista que comienza con la letra dada por el usuario.

    Parámetros:
    cadenas (list): Lista de cadenas.
    letra (str): La letra por la cual deben empezar las cadenas.
    """
    print(f"Cadenas que comienzan con la letra '{letra}':")
    for cadena in cadenas:
        if cadena.startswith(letra):
            print(cadena)

def contar_subcadenas(cadenas, subcadena):
    """
    Cuenta cuántas cadenas contienen la subcadena pedida por el usuario.

    Parámetros:
    cadenas (list): Lista de cadenas.
    subcadena (str): La subcadena que se busca en las cadenas.

    Retorna:
    int: El número de cadenas que contienen la subcadena.
    """
    contador = sum(1 for cadena in cadenas if subcadena in cadena)
    return contador

def encontrar_cadena_larga_corta(cadenas):
    """
    Encuentra la cadena más larga y la más corta de la lista.

    Parámetros:
    cadenas (list): Lista de cadenas.

    Retorna:
    tuple: La cadena más larga y la más corta.
    """
    cadena_larga = max(cadenas, key=len)
    cadena_corta = min(cadenas, key=len)
    return cadena_larga, cadena_corta

def verificar_mismo_objeto(cadenas, cadena1, cadena2):
    """
    Verifica si dos cadenas de la lista son el mismo objeto utilizando el operador is.

    Parámetros:
    cadenas (list): Lista de cadenas.
    cadena1 (str): La primera cadena.
    cadena2 (str): La segunda cadena.

    Retorna:
    bool: True si las dos cadenas son el mismo objeto, False en caso contrario.
    """
    return cadena1 is cadena2

def verificar_longitud_mayor_10(cadenas):
    """
    Verifica si alguna cadena de la lista tiene una longitud mayor que 10 utilizando un bucle for...else.

    Parámetros:
    cadenas (list): Lista de cadenas.
    """
    for cadena in cadenas:
        if len(cadena) > 10:
            print(f"La cadena '{cadena}' tiene más de 10 caracteres.")
            break
    else:
        print("Ninguna cadena tiene más de 10 caracteres.")

def main():
    """
    Función principal que ejecuta el menú y las funciones correspondientes.
    """
    cadenas = input("Ingrese una lista de cadenas separadas por espacios: ").split()

    while True:
        print("\nMenú de opciones:")
        print("1. Imprimir cadenas que comiencen con una letra específica")
        print("2. Contar cuántas cadenas contienen una subcadena")
        print("3. Encontrar la cadena más larga y la más corta de la lista")
        print("4. Verificar si dos cadenas son el mismo objeto")
        print("5. Verificar si alguna cadena tiene más de 10 caracteres")
        print("6. Salir")

        opcion = input("Elige una opción: ")

        if opcion == '1':
            # Opción 1: Imprimir cadenas que comienzan con una letra dada
            letra = input("Ingrese la letra por la cual deben comenzar las cadenas: ")
            imprimir_cadenas_por_letra(cadenas, letra)

        elif opcion == '2':
            # Opción 2: Contar cuántas cadenas contienen una subcadena
            subcadena = input("Ingrese la subcadena a buscar: ")
            contador = contar_subcadenas(cadenas, subcadena)
            print(f"{contador} cadena(s) contienen la subcadena '{subcadena}'.")

        elif opcion == '3':
            # Opción 3: Encontrar la cadena más larga y la más corta
            cadena_larga, cadena_corta = encontrar_cadena_larga_corta(cadenas)
            print(f"La cadena más larga es: '{cadena_larga}'")
            print(f"La cadena más corta es: '{cadena_corta}'")

        elif opcion == '4':
            # Opción 4: Verificar si dos cadenas son el mismo objeto
            cadena1 = input("Ingrese la primera cadena: ")
            cadena2 = input("Ingrese la segunda cadena: ")
            if verificar_mismo_objeto(cadenas, cadena1, cadena2):
                print("Las cadenas son el mismo objeto en memoria.")
            else:
                print("Las cadenas no son el mismo objeto en memoria.")

        elif opcion == '5':
            # Opción 5: Verificar si alguna cadena tiene más de 10 caracteres
            verificar_longitud_mayor_10(cadenas)

        elif opcion == '6':
            # Opción 6: Salir del programa
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida. Por favor, elija una opción del 1 al 6.")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
a
```
